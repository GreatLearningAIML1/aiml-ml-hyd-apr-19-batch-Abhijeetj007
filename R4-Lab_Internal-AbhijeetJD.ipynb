### 3. Standardize the data

```python
from scipy.stats import zscore
data1 = data1.apply(zscore)
data1.head()
```

### K - Means Clustering

### 4. Assign a dummy array called distortion

```python
distortions = []
```

You can later use this array to plot the elbow plot

### 5. Calculating distortion - C distance

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(data1)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

### 6. Plotting Elbow/ Scree Plot

```python


# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

Use Matplotlib to plot the scree plot - Note: Scree plot plots distortion vs the no of clusters

### Find out the optimal value of K

```python
# From the graph above, the optimal K Value can be determined as '5'
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
import numpy as np
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

### 9. Combine all three arrays a,b,c into a dataframe

```python
conc = np.concatenate((a,b,c), axis = 0)
dfh = pd.DataFrame(conc)
dfh.head()
```

### 10. Use scatter matrix to print all the 3 distributions

```python
import seaborn as sns
sns.pairplot(dfh, size = 7, aspect = 0.5, )
```

### 11. Find out the linkage matrix

Use ward as linkage metric and distance as Eucledian

```python
# from sklearn.preprocessing import StandardScaler
# from sklearn.cluster import AgglomerativeClustering 
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist 
# scaler = StandardScaler()
# scaler.fit(dfh)
# dfh = scaler.transform(dfh)
# dfh = pd.DataFrame(dfh)
# colnames = dfh.columns 
# model = AgglomerativeClustering(n_clusters=11, affinity='euclidean',  linkage='ward')
# model.fit(dfh)
Z = linkage(dfh, 'ward', metric = 'euclidean')
c, coph_dists = cophenet(Z , pdist(dfh))
print(c)
```

### 12. Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(10, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z)
plt.tight_layout()
```

### 13. Recreate the dendrogram for last 12 merged clusters 

```python
plt.figure(figsize=(10, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, p=12, truncate_mode='lastp')
plt.tight_layout()
```
